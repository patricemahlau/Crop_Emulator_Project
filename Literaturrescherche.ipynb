# Literatur Research until now 

- @heinickeGlobalGriddedCrop2022: **Evaluate Process Bases Crop Models**
    - "Process-based crop models are capable of simulating the impacts of extreme summer conditions. These models accurately reflect observed crop yield responses to heat and drought, although they tend to underestimate the magnitude of heat-related effects."
    - testing of 13 GGCms Models from the ISIMIP on different crops (maize,...)
        - Heatwaves: Models underestimate crop failure due to exposure error (timing) + general model perfomance (modeled biology, etc.)
            - perfomance depends on sexposure -> improving overlap with heatwaves would improve models
        - Droughts: No correlation of simulated exposure and model perfomance 
            - perfomance underestimation depends mainly on general model perfomance
            - models often expierence low precipitation but still fail to simulate crop failure


- @frankeGGCMIPhase22020: **Implement Crop Emulators for further usage**
    - input for the study:
        - training data comes from: GGCMI Phase 2 dataset (Global Gridded Crop Model Intercomparison Project)
    - 9 process-based crop models (e.g., LPJmL, pDSSAT, EPIC)
        - 5 crops: maize, rice, soybean, wheat (spring & winter)
        - A systematic dataset where key drivers are varied:
            1. CO₂ (C)
            2. Temperature (T)
            3. Water/precipitation (W)
            4. Nitrogen (N)
            → called CTWN space
    - Each model is run for hundreds of combinations (~756) of these variables across the globe.
    - What they build:
        - For each grid cell (0.5° resolution), crop, and model:
        -> a statistical function that maps (C, T, W, N) → yield
        - Functional form: mostly a 3rd-order polynomial regression
        - one statistical model per:
            - model: CARAIB, LPJmL,...
            - crop: maize, rice, soybean, wheat
            - grid cell: 0,5 degree

- Problem with @FrankeGCMIPhase22020: trained on climatological mean yield response

- data they used: 
The input and output data for the simulation of global gridded crop yield were obtained from the GGCMI phase 2 dataset, which includes gridded crop yield projections at 0.5◦ longitudinal–latitudinal resolution for maize, spring wheat, winter wheat, and rice (Franke et al., 2020b). The input data for the process-based simulations in GGCMI phase 2 included data for climate, soil, atmospheric CO2 concentration, and nitrogen fertilizer application rates. Baseline (19802010) climate inputs were used from the AgMIP ModernEra Retrospective Analysis for Research and Applications (AgMERRA) forcing dataset, including daily maximum and minimum temperatures, precipitation, and solar radiation (Ruane et al., 2015).
